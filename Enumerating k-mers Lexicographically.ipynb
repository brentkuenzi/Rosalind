**Given**: A collection of at most 10 symbols defining an ordered alphabet, and a positive integer nn (nâ‰¤10nâ‰¤10).

**Return**: All strings of length nn that can be formed from the alphabet, ordered lexicographically (use the standard order of symbols in the English alphabet).

```python
symbols = "A C G T"
integer = 2
answer = """AA
AC
AG
AT
CA
CC
CG
CT
GA
GC
GG
GT
TA
TC
TG
TT"""
```

```python
import itertools
def enumerate_kmers(string,length):
    test = string.replace(" ","")
    kmers_list = list(itertools.product(test,repeat = length))
    kmers = []
    for i in kmers_list:
        kmers.append("".join(i))
    return "\n".join(kmers)
```
